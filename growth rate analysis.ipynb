```python
# There are some errors in the data. Sometimes the areas get smaller after after more days (areas should only increase or stay the same)
# Usually this is due to measurement error between closely-sized images. I will deal with this afterwards.
# However, sometimes the measurements are due to more grievious scaling errors. 
# I want to flag these up so that I can go back and check my measurements. I will use a cut-off of 0.1 mm^2 difference

growth_data ['Area_Change'] = growth_data.groupby('ID')['Area'].transform(pd.Series.diff) 
growth_data[growth_data ['Area_Change']<-0.1 ]
```

```python
# I went back to my data and checked these values. I removed some from the dataset and remeasured some. 
# The cleaned data is now in a new worksheet in the original file

cleaned_growth_data = pd.read_excel('growth data compiled.xlsx','Cleaned data') 
```

```python
# Check there are no remaining errors >0.1 in the data
cleaned_growth_data ['Area_Change'] = cleaned_growth_data.groupby('ID')['Area'].transform(pd.Series.diff) 
cleaned_growth_data[cleaned_growth_data['Area_Change'] <-0.1] 
```

```python
# import the line names from a separate file
lines = pd.read_excel('growth data compiled.xlsx','Lines') 
lines.set_index('Number', inplace=True)
type(cleaned_growth_data.Line)
lines
```

```python
cleaned_growth_data['Line'] = lines.ix[cleaned_growth_data['ID']].values
cleaned_growth_data
```

```python
# Now I want to deal with the smaller measurement errors.  
# If the growth area decreases with increasing days, I will assign the previous day's area to the latter. 
# This should deal with any small discrepancies with measurements.
# first quantify the errors in the existing dataset:
cleaned_growth_data[cleaned_growth_data['Area_Change'] <0] 
```

```python
# create a function that will force the current item in a list to take the maximum of all of the previous items
# always ascending

def replace_descending(x):
    n_elements = len(x)
    for element_index in arange(n_elements):
        x.iloc[element_index] = max(x[:element_index+1])
    return x


cleaned_growth_data['Area']=cleaned_growth_data.groupby(['ID'])['Area'].apply(replace_descending) 
cleaned_growth_data
```

```python
# reassign values for area_change
cleaned_growth_data['Area_Change'] = cleaned_growth_data.groupby('ID')['Area'].transform(pd.Series.diff)
```

```python
# Check there are no remaining errors in the data
cleaned_growth_data ['Area_Change'] = cleaned_growth_data.groupby('ID')['Area'].transform(pd.Series.diff) 
cleaned_growth_data[cleaned_growth_data['Area_Change'] <0]
```

```python
# add '% increase' column to cleanaed_growth_data
cleaned_growth_data['% increase'] = cleaned_growth_data.groupby('ID')['Area_Change'].transform(pd.Series.cumsum) 
cleaned_growth_data['% increase'] = cleaned_growth_data['% increase']*100
cleaned_growth_data
```

```python
# remove the IDs for which there is not complete data (e.g. missing days). This seems to be messing up the seaborn tsplot
g = cleaned_growth_data.groupby('ID')
size = g.size()
f = size[size == 5]
only_complete = cleaned_growth_data[cleaned_growth_data['ID'].isin(f.index)]
```
